```python
import openpyxl

# Initialize Excel workbook and worksheet
workbook = openpyxl.Workbook()
worksheet = workbook.active

# Define column names
worksheet['A1'] = 'Name'
worksheet['B1'] = 'Priority'
worksheet['C1'] = 'Path Pattern'
worksheet['D1'] = 'Host Header'

# Initialize variables to store the current values
current_name = ''
current_priority = ''
current_path_pattern = ''
current_host_header = ''

# Read the text file line by line
with open('ggtr-alb.txt', 'r') as file:
    row = 2  # Start writing data from the second row
    for line in file:
        line = line.strip()  # Remove leading and trailing whitespace
        if line.startswith('#'):
            continue  # Skip lines starting with '#'
        if line.startswith('name'):
            worksheet.cell(row=row, column=1, value=current_name)
            worksheet.cell(row=row, column=2, value=current_priority)
            worksheet.cell(row=row, column=3, value=current_path_pattern)
            worksheet.cell(row=row, column=4, value=current_host_header)
            current_name = line.split('=')[1].strip()
        if line.startswith('priority'):
            current_priority = line.split('=')[1].strip()
        elif line.startswith('path_patterns'):
            current_path_pattern = line.split('=')[1].strip()
        elif line.startswith('host_headers'):
            current_host_header = line.split('=')[1].strip()
        elif line.startswith('https_listener_index') :
            # Write the values to the Excel worksheet
            worksheet.cell(row=row, column=1, value=current_name)
            worksheet.cell(row=row, column=2, value=current_priority)
            worksheet.cell(row=row, column=3, value=current_path_pattern)
            worksheet.cell(row=row, column=4, value=current_host_header)
            row += 1
            current_priority = ''
            current_path_pattern = ''
            current_host_header = ''
        else:
            continue
            



# Save the Excel file
workbook.save('output799.xlsx')


import pandas as pd

# Read the Excel file into a DataFrame
excel_file = "output799.xlsx"  # Update with the path to your Excel file
df = pd.read_excel(excel_file)

# Sort the DataFrame by the "priority" column
sorted_df = df.sort_values(by="Priority", ascending=True)

# Save the sorted DataFrame back to the same Excel file
sorted_excel_file = "sorted_output_new.xlsx"  # Update with your desired output file name
sorted_df.to_excel(sorted_excel_file, index=False)
print(f"Excel file sorted and saved: {sorted_excel_file}")



import pandas as pd

# Read the Excel file (replace 'your_file.xlsx' with the actual file path)
df = pd.read_excel('sorted_output_new.xlsx')

# Define the keys to search for
# key1_to_search = "*1223332212002.serviceurl.in"
# key2_to_search = "/api/stg/*"

input_url = input("Enter the input URL: ")

# Split the URL by "https://"
parts = input_url.split("https://")

if len(parts) > 1:
    # The first part is after "https://" and before the first "/"
    key1_to_search = parts[1].split("/")[0]

    # The second part is after "serviceurl.in/"
    key2_to_search = input_url.split("serviceurl.in")[1]

    print("Host_Header:", key1_to_search)
    print("Path Pattern:", key2_to_search)
else:
    print("Invalid input URL format.")


key2_to_search='"'+key2_to_search+'"'



# Initialize an empty list to store matching rows
matching_rows = []

# Iterate through each row in the DataFrame
for index, row in df.iterrows():
    # Check if the values are not NaN (float) before searching
    if not pd.isna(row['Host Header']) and not pd.isna(row['Path Pattern']):
        if key1_to_search in row['Host Header'] and key2_to_search in row['Path Pattern']:
            # Found a matching row
            matching_rows.append({
                'Row Number': index + 2,
                'Name': row['Name'],
                'Priority': row['Priority'],
                'Path Pattern': row['Path Pattern']
            })

# Print the matching rows
for match in matching_rows:
    print(f"URL found on row number {match['Row Number']} with Name='{match['Name']}', Priority={match['Priority']}, path pattern='{match['Path Pattern']}'")

```
